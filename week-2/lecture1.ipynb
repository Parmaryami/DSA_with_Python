# Copying lists
How can we make a copy of a list?

A slice creates a new (sub)list from an old one

Recall l[:k] is l[0:k], l[k:] is l[k:len(l)]

Omitting both end points gives a full slice
l[:] == l[0:len(l)]

To make a copy of a list use a full slice
list2 = list1[:]

```python
x = 5
y = x
x = 7

print(x,y)  # 7,5
print(x is y)
print(x == y)

list1 = [1,3,5,7]
list2 = [1,3,5,7]
list3 = list2

print(list1 == list2)  # True, values are same
print(list1 is list2)  # False, different objects   
print(list2 is list3)  # True, same object
list2[0] = 99
print(list2)  # [99,3,5,7]
print(list3)  # [99,3,5,7] because list3 references list2
print(list1)  # [1,3,5,7] unchanged
print(list1 == list2)  # False, values are different now
print(list1 is list2)  # False, different objects

# Copying a list
list4 = list1[:]  # list4 references list1
print(list4 is list1)  # False, different objects
print(list4 == list1)  # True, same values
print(list4)  # [1,3,5,7]
list4[0] = 100  
print(list4)  # [100,3,5,7]
print(list1)  # [1,3,5,7] unchanged
```
